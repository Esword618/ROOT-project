```c++
TFile *guan = new TFile("guan1.root","recreate");
TTree *ming = new TTree("ming", "This is my first tree");

int i = 1;
double a = 2.2;
double b = 0.0;

ming->Branch("i",&i,"i/I");
ming->Branch("a",&a,"a/D");
ming->Branch("b",&a,"b/D");
for(int j=0;j<100;j++){
    i=j;
    a=(double)i*0.1;
    b=a*a/3;
    ming->Fill();
}

ming->Write();
```
