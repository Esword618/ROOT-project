```python
from ROOT import TFile,TTree,TCanvas,TH1F
import ctypes
import numpy as np
```

```python
rootFilePath = "../rootdata/M00.root"
f = TFile(rootFilePath)
```

```python
f.ls()
```

```python
muTL = TH1F("muTL", "muTL", 40000, 0, 40) # mu track length --> m
muKE = TH1F("muKE", "muKE", 10000, 0, 10000) # mu kinetic energy --> GeV
mudepE = TH1F("mudepE", "mudepE", 4000, 0, 4000) # mu kinetic energy --> GeV
```

```python
outFilePath = "../rootdata/merged.root"
f_merged = TFile(outFilePath, "recreate")
t_merged = TTree("Merged", "production tree");
```

```python
Nmax = 1000
shape = (Nmax)
Nsignals = ctypes.c_int()
pdg = np.zeros(shape,dtype=np.intc)
x = np.zeros(shape,dtype=np.double)
y = np.zeros(shape,dtype=np.double)
z = np.zeros(shape,dtype=np.double)
t = np.zeros(shape,dtype=np.double)
Edep = np.zeros(shape,dtype=np.double)
QE = np.zeros(shape,dtype=np.double)
tcap = np.zeros(shape,dtype=np.double)
KE = np.zeros(shape,dtype=np.double)
```

```python
t_merged.Branch("Nsignals", Nsignals, "Nsignals/I")
t_merged.Branch("pdg", pdg, "pdg[Nsignals]/I")
t_merged.Branch("x", x, "x[Nsignals]/D")
t_merged.Branch("y", y, "y[Nsignals]/D")
t_merged.Branch("z", z, "z[Nsignals]/D")
t_merged.Branch("t", t, "t[Nsignals]/D")
t_merged.Branch("Edep", Edep, "Edep[Nsignals]/D")
t_merged.Branch("QE", QE, "QE[Nsignals]/D")
t_merged.Branch("tcap", tcap, "tcap[Nsignals]/D")
t_merged.Branch("KE", KE, "KE[Nsignals]/D")
```

```python
# 从 MuInitPosx 到 mustpz 这些数组中有些元素的主要是由 MuMult 决定。（注意）
N = 25
evtID, MuMult = ctypes.c_int(1), ctypes.c_int()
PDG = np.zeros((N),dtype=np.intc)
MuInitPosx = np.zeros((N),dtype=np.double)
MuInitPosy = np.zeros((N),dtype=np.double)
MuInitPosz = np.zeros((N),dtype=np.double)
MuInitPx = np.zeros((N),dtype=np.double)
MuInitPy = np.zeros((N),dtype=np.double)
MuInitPz = np.zeros((N),dtype=np.double)
MuInitKine = np.zeros((N),dtype=np.double)
TrackLengthInRock = np.zeros((N),dtype=np.double)
TrackLengthInVetoWater = np.zeros((N),dtype=np.double)
TrackLengthInCDWater = np.zeros((N),dtype=np.double)
TrackLengthInAcrylic = np.zeros((N),dtype=np.double)
TrackLengthInSteel = np.zeros((N),dtype=np.double)
TrackLengthInScint = np.zeros((N),dtype=np.double)
ELossInRock = np.zeros((N),dtype=np.double)
ELossInVetoWater = np.zeros((N),dtype=np.double)
ELossInCDWater = np.zeros((N),dtype=np.double)
ELossInAcrylic = np.zeros((N),dtype=np.double)
ELossInSteel = np.zeros((N),dtype=np.double)
ELossInScint = np.zeros((N),dtype=np.double)
MuExitPosx = np.zeros((N),dtype=np.double)
MuExitPosy = np.zeros((N),dtype=np.double)
MuExitPosz = np.zeros((N),dtype=np.double)
mustpMaterial = np.zeros((N),dtype=np.double)
mustpx = np.zeros((N),dtype=np.double)
mustpy = np.zeros((N),dtype=np.double)
mustpz = np.zeros((N),dtype=np.double)
```

```python
mu_tree.SetBranchAddress("evtID",evtID)
mu_tree.SetBranchAddress("MuMult",MuMult)
mu_tree.SetBranchAddress("PDG",PDG)
mu_tree.SetBranchAddress("MuInitPosx",MuInitPosx)
mu_tree.SetBranchAddress("MuInitPosy",MuInitPosy)
mu_tree.SetBranchAddress("MuInitPosz",MuInitPosz)
mu_tree.SetBranchAddress("MuInitPx",MuInitPx)
mu_tree.SetBranchAddress("MuInitPy",MuInitPy)
mu_tree.SetBranchAddress("MuInitPz",MuInitPz)
mu_tree.SetBranchAddress("MuInitKine",MuInitKine)
mu_tree.SetBranchAddress("TrackLengthInRock",TrackLengthInRock)
mu_tree.SetBranchAddress("TrackLengthInVetoWater",TrackLengthInVetoWater)
mu_tree.SetBranchAddress("TrackLengthInCDWater",TrackLengthInCDWater)
mu_tree.SetBranchAddress("TrackLengthInAcrylic",TrackLengthInAcrylic)
mu_tree.SetBranchAddress("TrackLengthInSteel",TrackLengthInSteel)
mu_tree.SetBranchAddress("TrackLengthInScint",TrackLengthInScint)
mu_tree.SetBranchAddress("ELossInRock",ELossInRock)
mu_tree.SetBranchAddress("ELossInVetoWater",ELossInVetoWater)
mu_tree.SetBranchAddress("ELossInCDWater",ELossInCDWater)
mu_tree.SetBranchAddress("ELossInAcrylic",ELossInAcrylic)
mu_tree.SetBranchAddress("ELossInSteel",ELossInSteel)
mu_tree.SetBranchAddress("ELossInScint",ELossInScint)
mu_tree.SetBranchAddress("MuExitPosx",MuExitPosx)
mu_tree.SetBranchAddress("MuExitPosy",MuExitPosy)
mu_tree.SetBranchAddress("MuExitPosz",MuExitPosz)
mu_tree.SetBranchAddress("mustpMaterial",mustpMaterial)
mu_tree.SetBranchAddress("mustpx",mustpx)
mu_tree.SetBranchAddress("mustpy",mustpy)
mu_tree.SetBranchAddress("mustpz",mustpz)
```

```python
nentries = mu_tree.GetEntries()
nentries
```

```python
# nentries
for i in range(100):
    mu_tree.GetEntry(i)
    print(f"Entry: {i}")
    print(f"MuMult: {MuMult.value}")
    # print(MuMult.value)
    for ii in range(MuMult.value):
        if TrackLengthInScint[ii]>0:
            muTL.Fill(TrackLengthInScint[ii]*1.e-3)
            muKE.Fill(MuInitKine[ii]*1.e-3)
            mudepE.Fill(-1.*ELossInScint[ii]*1.e-3)
            
            print(f"TrackLengthInScint:{TrackLengthInScint}")
            print(f"TrackLengthInScint[{i}]: ",TrackLengthInScint[ii]*1.e-3)
            print(f"MuInitKine:{MuInitKine}")
            print(f"MuInitKine[{i}]: ",MuInitKine[ii]*1.e-3)
            print(f"ELossInScint:{ELossInScint}")
            print(f"ELossInScint[{i}]: ",-1.*ELossInScint[ii]*1.e-3)
            print("----------------------------")
```

## mufastn

```python
mufastn_tree = f.Get("mufastn")
```

```python
mufastn_tree.Show(4)
```

```python
mufastn_tree.Print()
```

```python
shape = (10000)
evtID_n = ctypes.c_int()
mufastn_tree.SetBranchAddress("evtID",evtID_n)
stepNumber = ctypes.c_int()
mufastn_tree.SetBranchAddress("stepNumber",stepNumber)
fnPDG = np.zeros(shape,dtype=np.intc)
mufastn_tree.SetBranchAddress("fnPDG",fnPDG)
fnPositionx = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnPositionx",fnPositionx)
fnPositiony = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnPositiony",fnPositiony)
fnPositionz = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnPositionz",fnPositionz)
fnGlobalTime = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnGlobalTime",fnGlobalTime)
fnLocalTime = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnLocalTime",fnLocalTime)
fnTotalEnergy = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnTotalEnergy",fnTotalEnergy)
fnKineticEnergy = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnKineticEnergy",fnKineticEnergy)
fnEnergyDeposit = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnEnergyDeposit",fnEnergyDeposit)
fnQEnergyDeposit = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnQEnergyDeposit",fnQEnergyDeposit)
fnCaptime = np.zeros(shape,dtype=np.double)
mufastn_tree.SetBranchAddress("fnCaptime",fnCaptime)
```

```python
nentries = mufastn_tree.GetEntries()
nentries
```

```python
for i in range(nentries):
    # print("i:",i)
    mufastn_tree.GetEntry(i)
    if (stepNumber.value >= 1):
        tagMuFlag = False
        neuFlag = False
        for j in range(MuMult.value):
            trackInScint = TrackLengthInScint[j]
            if trackInScint>0:
                tagMuFlag = True
            else:
                tagMuFlag = False
        
        for k in range(stepNumber.value):
            if (fnCaptime[k]>0 and fnPDG[k] == 2112):
                neuFlag = True
        
        if tagMuFlag and neuFlag:
            print(f"evtID_n: {evtID_n.value}")
            t_PDG = 0
            t_x,t_y,t_z = 0.0,0.0,0.0
            t_t = 0.0
            t_edep = 0.0
            t_qe = 0.0
            t_cap = 0.0
            t_ke = 0
            t_win = 0
            Nsignals.value = 0
            steps = 0
            for m in range(Nmax):
                pdg[m] =0
                x[m] = 0
                y[m] = 0
                z[m] = 0
                t[m] = 0
                Edep[m] = 0
                QE[m] = 0
                tcap[m] = 0
                KE[m] = 0
            for stepN in range(stepNumber.value):
                if stepN == 0:
                    t_t = fnGlobalTime[stepN]
                    t_x = fnPositionx[stepN] * fnEnergyDeposit[stepN]
                    t_y = fnPositiony[stepN] * fnEnergyDeposit[stepN]
                    t_z = fnPositionz[stepN] * fnEnergyDeposit[stepN]
                    t_edep = fnEnergyDeposit[stepN]
                    t_qe = fnQEnergyDeposit[stepN]
                    t_ke = fnKineticEnergy[stepN]
                    steps += 1
                    t_PDG = 0
                    t_tcap = 0
                    if fnCaptime[stepN] > 0 and fnPDG[stepN] == 2112:
                        t_PDG = 2112
                        t_tcap = fnCaptime[stepN]
                else:
                    dt = fnGlobalTime[stepN] - t_t
                    if dt < 0:
                        print("Warning  ::: Call Jie : please check input root file")
                    if 0 < dt < t_win:
                        t_x += fnPositionx[stepN] * fnEnergyDeposit[stepN]
                        t_y += fnPositiony[stepN] * fnEnergyDeposit[stepN]
                        t_z += fnPositionz[stepN] * fnEnergyDeposit[stepN]
                        t_edep += fnEnergyDeposit[stepN]
                        t_qe += fnQEnergyDeposit[stepN]
                        t_ke += fnKineticEnergy[stepN]
                        steps += 1
                        if fnCaptime[stepN] > 0 and fnPDG[stepN] == 2112:
                            if t_PDG == 2112:
                                print("Warning  ::: Call Jie : please check input root file for double neutrons")
                            t_PDG = 2112
                            t_tcap = fnCaptime[stepN]
                    elif dt >= t_win:
                        pdg[Nsignals.value] = t_PDG
                        x[Nsignals.value] = t_x / t_edep
                        y[Nsignals.value] = t_y / t_edep
                        z[Nsignals.value] = t_z / t_edep
                        Edep[Nsignals.value] = t_edep
                        QE[Nsignals.value] = t_qe
                        t[Nsignals.value] = t_t * 1.e-3
                        tcap[Nsignals.value] = t_tcap
                        KE[Nsignals.value] = t_ke / steps
                        steps = 0
                        Nsignals.value += 1

                        t_PDG = 0
                        t_tcap = 0
                        t_t = fnGlobalTime[stepN]
                        t_x = fnPositionx[stepN] * fnEnergyDeposit[stepN]
                        t_y = fnPositiony[stepN] * fnEnergyDeposit[stepN]
                        t_z = fnPositionz[stepN] * fnEnergyDeposit[stepN]
                        t_edep = fnEnergyDeposit[stepN]
                        t_qe = fnQEnergyDeposit[stepN]
                        t_ke = fnKineticEnergy[stepN]
                        steps = 1
                        if fnCaptime[stepN] > 0 and fnPDG[stepN] == 2112:
                            t_PDG = 2112
                            t_tcap = fnCaptime[stepN]

                if stepN == stepNumber.value - 1:
                    pdg[Nsignals.value] = t_PDG
                    x[Nsignals.value] = t_x / t_edep
                    y[Nsignals.value] = t_y / t_edep
                    z[Nsignals.value] = t_z / t_edep
                    Edep[Nsignals.value] = t_edep
                    QE[Nsignals.value] = t_qe
                    t[Nsignals.value] = t_t * 1.e-3
                    tcap[Nsignals.value] = t_tcap
                    KE[Nsignals.value] = t_ke / steps
                    steps = 0
                    Nsignals.value += 1
f_merged.cd()
muTL.Write()
muKE.Write()
mudepE.Write()
t_merged.Write()
f_merged.Close()
```
