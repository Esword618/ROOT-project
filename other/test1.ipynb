```python
from ROOT import TCanvas, TGraph, gStyle,TH2D
import math

# 创建一个画布
c2 = TCanvas("c2", "Cardioid Example", 800, 800)

# 设置画布样式
gStyle.SetOptStat(0)

# 创建一个 TGraph 对象
cardioid = TGraph()

# 定义直角坐标方程
def cardioid_equation(t, a):
    x = a * math.cos(t)
    y = a * math.sin(t) * math.cos(t)
    return x, y

# 添加点到 TGraph
n_points = 1000
a = 1  # 尺度因子
for i in range(n_points):
    t = 2 * math.pi * i / n_points
    x, y = cardioid_equation(t, a)
    cardioid.SetPoint(i, x, y)

# 设置图形样式
cardioid.SetLineColor(2)  # 红色

# 绘制图形
cardioid.Draw("AL")  # AL 表示绘制带有线条的图形

# 显示画布
c2.Draw()
```

```python
# 创建一个画布
c3 = TCanvas("c3", "Heart-shaped Example", 800, 800)

# 设置画布样式
gStyle.SetOptStat(0)

# 创建一个 TGraph 对象
heart = TGraph()

# 定义直角坐标方程
def heart_equation(t, a):
    x = a * (16 * math.sin(t)**3)
    y = a * (13 * math.cos(t) - 5 * math.cos(2 * t) - 2 * math.cos(3 * t) - math.cos(4 * t))
    return x, y

# 添加点到 TGraph
n_points = 1000
a = 1  # 尺度因子
for i in range(n_points):
    t = 2 * math.pi * i / n_points
    x, y = heart_equation(t, a)
    heart.SetPoint(i, x, y)

# 设置图形样式
heart.SetLineColor(2)  # 红色

# 绘制图形
heart.Draw("AL")  # AL 表示绘制带有线条的图形

# 显示画布
c3.Draw()
```

```python
# 创建一个画布
c4 = TCanvas("c4", "Heart-shaped Example", 800, 800)

# 设置画布样式
gStyle.SetOptStat(0)

# 创建一个 TH2D 对象
heart_hist = TH2D("heart_hist", "Heart-shaped Example", 100, -2, 2, 100, -2, 2)

# 定义直角坐标方程
def heart_equation(t, a):
    x = a * (16 * math.sin(t)**3)
    y = a * (13 * math.cos(t) - 5 * math.cos(2 * t) - 2 * math.cos(3 * t) - math.cos(4 * t))
    return x, y

# 添加点到 TH2D
n_points = 1000
a = 0.1  # 尺度因子
for i in range(n_points):
    t = 2 * math.pi * i / n_points
    x, y = heart_equation(t, a)
    heart_hist.Fill(x, y)

# 设置图形样式
heart_hist.SetLineColor(2)  # 红色
# 绘制图形
heart_hist.Draw("COLZ")  # COLZ 表示使用颜色绘制

# 显示画布
c4.Draw()
```
